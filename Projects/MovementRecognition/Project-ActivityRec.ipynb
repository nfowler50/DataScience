```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix

from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
from sklearn.svm import LinearSVC
from sklearn.preprocessing import MinMaxScaler
```

```python
features=pd.read_csv('features.txt',delim_whitespace=True,header=None)
X=pd.read_csv('train/X_train.txt',delim_whitespace=True,header=None)
X.columns=features
X[:5]
```

```python
X.info()
```

```python
y=pd.read_csv('train/y_train.txt',delim_whitespace=True,header=None)
y[:5]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
lr=LogisticRegression()
lr.fit(X_train,y_train)
print('train score: {}'.format(lr.score(X_train,y_train)))
print('test score: {}'.format(lr.score(X_test,y_test)))
print(confusion_matrix(y_test,lr.predict(X_test)))
```

```python
SGD=SGDClassifier()
SGD.fit(X_train,y_train)
print('train score: {}'.format(SGD.score(X_train,y_train)))
print('test score: {}'.format(SGD.score(X_test,y_test)))
print(confusion_matrix(y_test,SGD.predict(X_test)))
```

```python
SVC=LinearSVC()
SVC.fit(X_train,y_train)
print('train score: {}'.format(SVC.score(X_train,y_train)))
print('test score: {}'.format(SVC.score(X_test,y_test)))
print(confusion_matrix(y_test,SVC.predict(X_test)))
```

```python
scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(X)
Xs_train,Xs_test,ys_train,ys_test=train_test_split(X_scaled,y)
```

```python
SVC=LinearSVC()
SVC.fit(Xs_train,ys_train)
print('train score: {}'.format(SVC.score(Xs_train,ys_train)))
print('test score: {}'.format(SVC.score(Xs_test,ys_test)))
print(confusion_matrix(ys_test,SVC.predict(Xs_test)))
```
