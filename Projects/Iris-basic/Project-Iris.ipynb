```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys, csv, os

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
```

```python
data=pd.read_csv('iris.data', names=['sepal_len','sepal_wid','petal_len','petal_wid','class'])
data[:5]
```

```python
data.columns
```

```python
y=list(data['class'])
y[:5]
```

```python
X=np.array(data[['sepal_len','sepal_wid','petal_len','petal_wid']])
```

```python
X[:5]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
clf=KNeighborsClassifier(n_neighbors=3)
```

```python
clf.fit(X_train,y_train)
print(clf.score(X_train,y_train))
print(clf.score(X_test,y_test))

pred_knn=clf.predict(X_test)
confusion=confusion_matrix(y_test,pred_knn)
print(confusion)
```

```python
svm=LinearSVC().fit(X_train,y_train)
print(svm.score(X_train,y_train))
print(svm.score(X_test,y_test))

pred_linSVC=svm.predict(X_test)
confusion=confusion_matrix(y_test,pred_linSVC)
print(confusion)
```

```python
param_grid=[{'kernel':['rbf'],
            'C':[0.001,0.01,0.1,1,10,100], 'gamma':[0.001,0.01,0.1,1,10,100]},
           {'kernel':['linear'],
           'C':[0.001,0.01,0.1,1,10,100]}]
```

```python
grid_search=GridSearchCV(SVC(),param_grid,cv=5)
grid_search.fit(X_train,y_train)
```

```python
print(grid_search.best_params_)
print(grid_search.best_score_)
```

```python
svm_opt=SVC(C=100,kernel='linear').fit(X_train,y_train)
print(svm.score(X_train,y_train))
print(svm.score(X_test,y_test))

pred_SVC=svm_opt.predict(X_test)
confusion=confusion_matrix(y_test,pred_SVC)
print(confusion)
```
