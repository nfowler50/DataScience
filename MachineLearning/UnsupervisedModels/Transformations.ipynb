```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVC

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

cancer=load_breast_cancer()
X_train, X_test, y_train, y_test=train_test_split(cancer.data,cancer.target, random_state=0)

print(X_train.shape, '\n', X_test.shape)
```

```python
scaler=MinMaxScaler()

scaler.fit(X_train)
```

```python
X_train_scaled=scaler.transform(X_train)
print('transformed shape: {}'.format(X_train_scaled.shape))
print('og min: \n {}'.format(X_train.min(axis=0)))
print('og max: \n {}'.format(X_train.max(axis=0)))
print('scaled min: \n {}'.format(X_train_scaled.min(axis=0)))
print('scaled max: \n {}'.format(X_train_scaled.max(axis=0)))
```

```python
#must transform test set before running svm

X_test_scaled=scaler.transform(X_test)

print('scaled min: \n {}'.format(X_test_scaled.min(axis=0)))
print('scaled max: \n {}'.format(X_test_scaled.max(axis=0)))
```

```python
#try sci kits StandardScaler - sets mean to 0 and variance to 1
# RobustScaler - uses median and quartiles: less impact by outliers
# MinMaxScaler used here sets min 0 max 1
# Normalizer sets euclidean feature vector to 1
```

```python
#either fit then transform or try fit_transform

from sklearn.preprocessing import StandardScaler
stdscaler=StandardScaler()
scaler = MinMaxScaler()
X_train_scaled=scaler.fit(X_train).transform(X_train)

# OR

X_test_scaled_d=scaler.fit_transform(X_train)
```

```python
#lets look at the effects of scaling on SVM
```

```python
svm=SVC(C=100)
svm.fit(X_train,y_train)
print('prescaling test accuracy: {}'.format(svm.score(X_test,y_test)))

svm.fit(X_train_scaled, y_train)
print('scaled test accuracy: {}'.format(svm.score(X_test_scaled,y_test)))
```
