```python
data=pd.DataFrame({'food':['bacon','pulled_pork','bacon','Pastrami','corned_beef','bacon','pastrami','honey_ham','nova_lox'],
                  'ounces':[4,3,12,6,7.5,8,3,5,6]})
data
```

```python
meat_to_animal={
    'bacon':'pig',
    'pulled_pork':'pig',
    'pastrami':'cow',
    'corned_beef':'cow',
    'honey_ham':'pig',
    'nova_lox':'salmon'
}
```

```python
data['food']=data['food'].str.lower()
data
```

```python
data['animal']=data.food.map(meat_to_animal)
data
```
